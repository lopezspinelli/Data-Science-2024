######Visualización. Graficos de dispersion

Los gráficos de dispersión tambien conocidos como scatter plots son herramientas visuales que se utilizan pra a muchos propositos.

-Visualización de Relaciones Bivariadas: Los scatter plots permiten observar la relación entre dos variables numéricas. Cada punto en el gráfico representa un par de valores correspondientes a las dos variables.

 -Identificación de Correlaciones: Los patrones en la distribución de los puntos pueden sugerir una correlación positiva, negativa o ninguna correlación entre las variables.

 -Detección de Outliers: Los puntos que se encuentran lejos del patrón general del resto de los datos pueden ser identificados fácilmente como valores atípicos o outliers.

 -Análisis de Tendencias: Los scatter plots pueden revelar tendencias generales en los datos, y mediante la adición de líneas de tendencia (regresión), estas tendencias pueden ser cuantificadas.

-Comparación de Grupos: Mediante el uso de diferentes colores o formas para los puntos, los scatter plots pueden compararse entre diferentes categorías o grupos dentro del mismo conjunto de datos.

-Visualización de Distribuciones: Ayuda a visualizar la distribución y dispersión de los datos, proporcionando una idea de la variabilidad y la concentración de los datos.

######Calculo de la desviación estandar

El cálculo de la desviación estándar es fundamental en estadística y análisis de datos. Con el calculo de la desviación podemos:

-Medida de Dispersión:
La desviación estándar mide la dispersión de los datos respecto a la media. Cuanto mayor sea la desviación estándar, mayor será la variabilidad de los datos.

-Comprender la Variabilidad:
Proporciona una idea clara de cómo los datos se distribuyen alrededor de la media. Es útil para entender si los datos están muy dispersos o concentrados cerca de la media.

-Identificación de Outliers:
Los valores que se encuentran a más de unas cuantas desviaciones estándar de la media pueden considerarse outliers o valores atípicos.

-Comparación de Conjuntos de Datos:
Facilita la comparación de la dispersión entre diferentes conjuntos de datos. Dos conjuntos de datos con la misma media pueden tener diferente variabilidad, lo cual se refleja en sus desviaciones estándar.

-Análisis de Riesgo y Desempeño:
En finanzas, la desviación estándar se utiliza para medir la volatilidad de un activo o una cartera. Un activo con una alta desviación estándar se considera más riesgoso.

#####Random Forest Regressor

Utilizamos el modelo de aprendizaje automático Random Forest Regressor, que es capaz de manejar con robustez la gran cantidad de valores atípicos que contiene el dataset.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

y=data['Close']

columns=['Open', 'High', 'Low','Volume']

X=data[columns]
X.head()

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
rf_val_predictions = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)

print("Validation MAE for Random Forest Model: {:,.6f}".format(rf_val_mae))
```

Teniendo en cuenta la media de la cotización de Dogecoin de los últimos 6 meses, y el Error Medio Absoluto obtenido, verificamos que la predicción tiene un margen de error de +/- 1,81%

```python
#Dividimos el MAE por la media y multiplicamos por 100 para obtener el margen de error
print('El margen de error es de:',(rf_val_mae/media)*100, '%')
```
